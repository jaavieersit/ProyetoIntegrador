```python

# archivos csv
import pandas as pd
```

2. CON LA LIBRERIA PANDAS LEEMOS EL ARCHIVO CSV

```python
df = pd.read_csv("0.dataset_diabetes.csv")

```

3.TRADUCIMOS LAS COLUMNAS DEL DATASET

```python
df.columns = ["Genero", "Edad", "Hipertencion", "Cardiopatia", "Historial_fumador", "IMC", "Nivel_HnA1c", "Glucosa_en_sangre", "Diabetes"]
df
```

4.SE ELIMINARAN TODAS LAS FILAS QUE CONTENGAN LA INFORMACION "NO INFO" DEL "HISTORIAL DE FUMADOR"

```python
df = df[df["Historial_fumador"] != "No Info"]
```


5.ELIMINAMOS EL INDICE ANTERIOR Y CREAMOS UNO DE FORMA ASCENDENTE

```python
df = df.reset_index(drop=True)
```


6.TRADUCCION DE LOS STRINGS DEL DATASET

```python
df["Genero"] = df["Genero"].replace({"Male": "Masculino", "Female": "Femenino"})
df["Historial_fumador"] = df["Historial_fumador"].replace({"never": "Nunca", "current": "Actualmente", "former": "Anteriormente", "ever": "Alguna vez", "not current": "Anteriormente"})
```

```python


# DefiNINIENDO LOS RANGOS DE EDAD:
# Rangos de edad: JOVENES[0, 30), ADULTOS[30, 60), PERSONAS MAYORES[60, 150)
edades = [0, 30, 60, 150] 

# Crea una nueva columna de grupo de edad en el dataset
df['Grupo de Edad'] = pd.cut(df['Edad'], bins=edades, labels=['JÃ³venes', 'Adultos', 'Personas Mayores'])

# Calcula la distribuciÃ³n de la diabetes en cada grupo de edad
distribucion = df.groupby('Grupo de Edad')['Diabetes'].value_counts().unstack()


```
